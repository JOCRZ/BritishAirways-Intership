```python


DataSet = pd.DataFrame()

for j in range (1,184):

    url = 'https://www.airlinequality.com/airline-reviews/british-airways/page/{}/?sortby=post_date%3ADesc&pagesize=20'.format(j)
    webpage = requests.get(url).text

    soup = BeautifulSoup(webpage)

    date = []
    rating = []
    title = []
    description = []

    for i in range(10):
        date.append([i for i in soup.findAll('h3', class_='text_sub_header userStatusWrapper')][i].text.strip())
        rating.append([i for i in soup.findAll('div',class_='rating-10')][i].text.strip())
        title.append([i for i in soup.findAll('h2',class_='text_header')][i].text.strip())
        description.append([i for i in soup.findAll('div',class_='text_content')][i].text.strip())

    dictionary = {'Date':date,'Rating':rating,'Title':title,'Description':description}
    df = pd.DataFrame(dictionary)

    DataSet = pd.concat([DataSet, df], ignore_index=True)
```

```python
import re

def date_extraction(string):
    date_pattern = r"\d{1,2}(\w{2})? [A-Za-z]+ \d{4}"
    matches = re.search(date_pattern, string)
    if matches:
        extracted_date = matches.group()
        return extracted_date

df['Date'] = df['Date'].apply(date_extraction)


```

```python
df['Date'] = df['Date'].str.replace(r'(th|nd|rd|st)', '', regex=True)
```

```python
from datetime import datetime
```

```python
def date_convertion(string):
    date_format = "%d %B %Y"
    date = datetime.strptime(string,date_format)
    return date
    
df['Date'] = df['Date'].apply(date_convertion)
```

```python
def convert_decimal(string):

    parts = string.split('/')
    if len(parts) == 2:
        numerator = float(parts[0])
        denominator = float(parts[1])
        
        decimal_value = numerator / denominator
        return float(decimal_value)
    

df['Rating'] = df['Rating'].apply(convert_decimal)
```
